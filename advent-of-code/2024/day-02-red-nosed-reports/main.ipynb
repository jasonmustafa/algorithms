```python
from copy import deepcopy
```

```python
### Part 1 ###

reports = []

with open("input.txt", "r") as file:
    for report in file:
        reports.append(report.split())

reports = [[int(level) for level in report] for report in reports]
```

```python
num_safe = 0

# loop through reports
for report in reports:
    is_safe = True

    is_increasing = report[0] < report[1]
    is_decreasing = report[0] > report[1]
    
    if not is_increasing and not is_decreasing:
        is_safe = False
        continue

    # loop through all levels in report
    for i in range(1, len(report)):
        curr_level = report[i]
        prev_level = report[i - 1]

        # check increasing
        if is_increasing:
            if curr_level <= prev_level:
                is_safe = False
                break

        # check decreasing
        if not is_increasing:
            if curr_level >= prev_level:
                is_safe = False
                break

        # check difference
        diff = abs(curr_level - prev_level)
        if diff < 1 or diff > 3:
            is_safe = False
            break
    
    if is_safe:
        num_safe += 1

print(num_safe)
```

```python
### PART 2 ###

num_safe = 0

# loop through reports
for report in reports:
    # create each combo with one level removed
    combos = []
    for i in range(len(report)):
        combo = deepcopy(report)
        level = report[i]
        del combo[i]
        combos.append(combo)

    for combo in combos:
        is_safe = True

        is_increasing = combo[0] < combo[1]
        is_decreasing = combo[0] > combo[1]
        
        if not is_increasing and not is_decreasing:
            is_safe = False
            continue

        # loop through all levels in report
        for i in range(1, len(combo)):
            curr_level = combo[i]
            prev_level = combo[i - 1]

            # check increasing
            if is_increasing:
                if curr_level <= prev_level:
                    is_safe = False
                    break

            # check decreasing
            if is_decreasing:
                if curr_level >= prev_level:
                    is_safe = False
                    break

            # check difference
            diff = abs(curr_level - prev_level)
            if diff < 1 or diff > 3:
                is_safe = False
                break
    
        if is_safe:
            num_safe += 1
            break

print(num_safe)
```
